# 1. Import Dependencies

```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

#Basic Processing

```python
df = pd.read_csv('data/processed/ChurnModelling_Encoded.csv')
df.head()
```

```python
df['Tenure'].max()
```

```python
columns_need_to_scale = ['Age', 'Tenure', 'Balance', 'EstimatedSalary']

for col in columns_need_to_scale:
    scaler = StandardScaler()
    df[col] = scaler.fit_transform(df[col].values.reshape(10000, 1))
    
df.head()
```

```python
df.to_csv('data/processed/ChurnModelling_Final.csv', index=False)
```
