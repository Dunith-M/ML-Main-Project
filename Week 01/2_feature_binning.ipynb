#Binning using Credit Score ,     we can also use age , salary or any numerical one


```python
"""
Credit Score
350 - 580 => 'poor '
580 - 670 => 'average '
670 - 740 => 'good '
740 - 800 => 'very good '
800 - 850 => 'excellent'
"""
```

```python
def customer_binning_credit_score(score):
    if score < 580:
        return 'poor'
    if score < 670:
        return 'average'
    if score < 740:
        return 'good'
    if score < 800:
        return 'very good'
    if score <= 850:
        return 'excellent'
    else:
        assert True, "Invalid credit score"
        
df['CreditScoreBins'] = df['CreditScore'].apply(customer_binning_credit_score)
df.head()
```
