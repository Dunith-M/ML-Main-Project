```python
df = pd.read_csv('data/processed/ChurnModelling_Binning_Applied.csv')
df.head()
```

# 2. Encode Ordinal Variables

#Norminal Variables

```python
norminal_features = ['Geography', 'Gender']

geograpgy_dummies = pd.get_dummies(df['Geography'], prefix='Geography')
gender_dummies = pd.get_dummies(df['Gender'], prefix='Gender')

df_encoded = pd.concat([df, geograpgy_dummies], axis=1)
del df_encoded['Geography']

df_encoded = pd.concat([df_encoded, gender_dummies], axis=1)
del df_encoded['Gender']

df_encoded
```

#Encode Ordinal Variables

```python
encode_dict_creditscore = {
                            'poor': 0,
                            'average': 1,
                            'good': 2,
                            'very good': 3,
                            'excellent': 4
}

df_encoded['CreditScoreBins'] = df_encoded['CreditScoreBins'].map(encode_dict_creditscore)
df_encoded.head(10)
```
